```python
def model_check(model):
    try:
        model.fit(x_train, y_train)

        y_pred_train = model.predict(x_train)
        y_pred_test = model.predict(x_test)


        print('===============','cross_score','===============')
        kfold = KFold(n_splits=5)
        cr_score = cross_val_score(model, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
        print(cr_score)
        print(cr_score.mean())


        print('===============','r2_score','===============')
        print(r2_score(y_train, y_pred_train))
        print(r2_score(y_test, y_pred_test))

        print('===============','mse','===============')
        print(mean_squared_error(y_train, y_pred_train))
        print(mean_squared_error(y_test, y_pred_test))


        print('===============','Rmse','===============')
        print(np.sqrt(mean_squared_error(y_train, y_pred_train)))
        print(np.sqrt(mean_squared_error(y_test, y_pred_test)))


        print('===============','mae','===============')
        print(mean_absolute_error(y_train, y_pred_train))
        print(mean_absolute_error(y_test, y_pred_test))

        print('===============','predicted','===============')
        result = pd.DataFrame({'real':y_test, 'pred':y_pred_test})
        result['diff'] = result['real'] - result['pred']
        sns.scatterplot(data=result, x='real', y='pred')
        plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
        plt.show()

        print('===============','residual','===============')
        sns.scatterplot(data=result, x='pred', y='diff')
        plt.axhline(y=0, linestyle='--', color='red')
        plt.show()

        print('===============','features_importances','===============')
        top20 = pd.Series(model.feature_importances_, index= x_train.columns)
        top20 = top20.sort_values(ascending=False)
        sns.barplot(x=top20, y=top20.index, color='orange')
        plt.show()
        
        plt.plot(top20.cumsum(), marker='o')
        plt.xticks(rotation=45)
        plt.grid(axis='y')
        plt.show()
    except Exception as e:
        print(f" generated error: {e}")
        pass
```

# Models Result Comapreing

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score


n = x_test.shape[0]
p = x_test.shape[1]
n, p

models = ['lr', 'rfr', 'xgb_r', 'lgb_r', 'tf_model', 'mlp_regressor']
results = []

def adjusted_r2(r_square, n, p):
    return 1 - ((1 - r_square) * (n - 1) / (n - p - 1))

n = x_test.shape[0]
p = x_test.shape[1]  

for model_name in models:
    model = globals()[model_name]
    
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test).flatten()  
    
    r2 = r2_score(y_test, y_pred)
    mae_val = mean_absolute_error(y_test, y_pred)
    mse_val = mean_squared_error(y_test, y_pred)
    rmse_val = np.sqrt(mse_val)
    mape_val = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    mpe_val = np.mean(y_test - y_pred)
    adj_r2_val = adjusted_r2(r2, n, p)
    exp_var_score = explained_variance_score(y_test, y_pred)
    
    results.append({
        'model': model_name,
        'R2_score': r2,
        'mse': mse_val,
        'rmse': rmse_val,
        'mae': mae_val,
        'mape': mape_val,
        'mpe': mpe_val,
        'explained_variance': exp_var_score
    })

models_summary = pd.DataFrame(results)
models_summary.round(2)
```
